# JIT Compile

```
import torch
from TripletFace.tripletface.core.model import Encoder

model = Encoder(64)
weigths = torch.load( '/content/TripletFace/model/model.pt' )['model']
model.load_state_dict( weigths )
jit_model = torch.jit.trace( model, torch.rand(8, 3, 3, 3) )
torch.jit.save( jit_model, '/content/jit/model1.pt' )
```
